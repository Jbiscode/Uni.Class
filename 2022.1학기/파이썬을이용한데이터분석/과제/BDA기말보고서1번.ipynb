### 1-(1)    월별 주문횟수

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/batang.ttc").get_name()
rc('font',family=font_name)

df = pd.read_excel('C:/대학수업/2022.1학기/파이썬을이용한데이터분석/data/sales.xlsx')

df['일자'] = pd.to_datetime(df['일자'])
df['월별'] = df['일자'].dt.to_period('M')
M_count = pd.crosstab(index = df['월별'],columns = '주문횟수')

M_count.plot(kind='bar')
plt.title('월별 주문횟수')
```

### 1-(3) 월별 주문액 합계

```python
import pandas as pd
from matplotlib import font_manager,rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/batang.ttc").get_name()
rc('font',family=font_name)

df = pd.read_excel('C:/대학수업/2022.1학기/파이썬을이용한데이터분석/data/sales.xlsx')

df['일자'] = pd.to_datetime(df['일자'])
df['월별'] = df['일자'].dt.to_period('M')

M_sum = df.pivot_table(index='월별',values='주문액',aggfunc='sum')


M_sum.plot(kind='bar')
plt.title('월별 주문액 합계')
```
