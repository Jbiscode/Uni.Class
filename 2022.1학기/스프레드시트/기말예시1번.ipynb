```python
import pandas as pd
df = pd.read_excel('C:/대학수업/2022.1학기/스프레드시트/data/apt_price_v1.xlsx')
data = df.copy()

def calc_pyung(x):
    if x<=50:
        size = '소형'
    elif x<=90:
        size = '중형'
    elif x<=120:
        size = '대형'
    else:
        size = '특대형'
    return size
def calc_price(x):
    if x<=30000:
        price = '저가'
    elif x<=70000:
        price = '중저가'
    elif x<=150000:
        price = '고가'
    else:
        price = '특고가'
    return price
data['평형'] = data['전용면적'].apply(calc_pyung)
data['가격대'] = data['거래금액'].apply(calc_price)
table_1 = pd.crosstab(index=data['가격대'],columns='가구수')
table_1['비중'] = pd.crosstab(index=data['가격대'],columns='비중',normalize='all')
table_2 = pd.crosstab(index=data['평형'],columns=data['가격대'],margins=True)

print(table_1,'\n',table_2)
```
