```python
import pandas as pd

df = pd.read_excel('C:/대학수업/2022.1학기/스프레드시트/중간데이터/restaurant.xlsx')
df.head(3)
```

```python
temp = pd.DataFrame()
temp['시'] = df['소재지'].str.split(expand=True)[1]
temp['동'] = df['소재지'].str.split(expand=True)[2]
temp
```

```python
df.drop('소재지',axis=1,inplace=True)
data = pd.concat([df,temp],axis=1)
data
```
